```python
df_clean["tenure_group"] = pd.cut(df_clean["tenure_months"], bins=[-1, 12, 24, 48, 72], labels=["0-12 Months", "13-24 Months", "25-48 Months", "49-72 Months"])
```

```python
df_clean["tenure_group"].value_counts().sort_index()
```

```python
df_clean["monthly_charge_group"] = pd.cut(
    df_clean["monthly_charges"],
    bins=[0, 35, 70, 120],
    labels=["Low Monthly Charge", "Medium Monthly Charge", "High Monthly Charge"]
)
```

```python
df_clean["monthly_charge_group"].value_counts().sort_index()
```

```python
df_clean["cltv_group"] = pd.qcut(
    df_clean["cltv"],
    q=3,
    labels=["Low CLTV", "Medium CLTV", "High CLTV"]
)
```

```python
df_clean["cltv_group"].value_counts().sort_index()
```
